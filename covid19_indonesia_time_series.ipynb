```python
import pandas as pd

# Baca file CSV ke dalam DataFrame
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')

# Mengecek apakah ada nilai NaN atau kosong di setiap baris pada tiap kolom
baris_dengan_nan = df.isna().any(axis=1)

# Menampilkan baris yang mengandung nilai NaN atau kosong
print(df[baris_dengan_nan])

# Mengganti nilai NaN dengan "0"
df.fillna(0, inplace=True)

# Menghilangkan kolom 'City or Regency'
df.drop(columns=['City or Regency'], inplace=True)

# Menghitung kolom-kolom yang masih memiliki nilai NaN
kolom_nan = df.columns[df.isna().any()].tolist()
print("Kolom dengan nilai NaN:", kolom_nan)

# Menyimpan DataFrame yang telah dimodifikasi ke dalam file CSV baru jika diperlukan
df.to_csv('covid19_indonesia_time_series_cleaned.csv', index=False)
```
